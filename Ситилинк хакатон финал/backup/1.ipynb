```python
import json

df = {}
for name in ['backup_71.txt', 'backup_268.txt',
             'backup_202.txt', "backup_91.txt",
             "backup_40.txt", "backup_5.txt",
             'backup_146.txt', "backup_178.txt"]:
    with open(path + "/" + name) as f:
        df.update(json.load(f))
```

```python
import pandas as pd

main_df = pd.DataFrame.from_dict(df, orient="index")
main_df.columns = ["hockey", "people", "maсhine"]
main_df = main_df.reset_index()
main_df["people"] = main_df["people"].astype(int)
main_df["date"] = main_df["index"].apply(lambda x: pd.to_datetime(x[:-4]))

main_df["maсhine"].iloc[463] = False
main_df["hockey"].iloc[463] = True

main_df["people"].iloc[238] = 0
main_df["hockey"].iloc[238] = True

main_df["people"].iloc[375] = 27

main_df["people"].iloc[391] = 0
main_df["hockey"].iloc[391] = True


main_df["maсhine"].iloc[611] = True
```
